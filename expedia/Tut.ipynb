```python
def accuracy(predictions, target_set):
    target = np.array([[l] for l in target_set["hotel_cluster"]])
    return metrics.mapk(target, predictions, k=5)    
```

```python
def make_key(items):
    return "_".join([str(i) for i in items])

def model1_train(match_cols, train_set):
    cluster_cols = match_cols + ['hotel_cluster']
    groups = train_set.groupby(cluster_cols)
    top_clusters = {}
    for name, group in groups:
        clicks = len(group.is_booking[group.is_booking == False])
        bookings = len(group.is_booking[group.is_booking == True])
    
        score = bookings + .15 * clicks
    
        clus_name = make_key(name[:len(match_cols)])
        if clus_name not in top_clusters:
            top_clusters[clus_name] = {}
        top_clusters[clus_name][name[-1]] = score

    cluster_dict = {}
    for n in top_clusters:
        tc = top_clusters[n]
        top = [l[0] for l in sorted(tc.items(), key=operator.itemgetter(1), reverse=True)[:5]]
        cluster_dict[n] = top

    return cluster_dict

def model1_run(match_cols, cluster_dict,valid_set):
    preds = []
    for index, row in valid_set.iterrows():
        key = make_key([row[m] for m in match_cols])
        if key in cluster_dict:
            preds.append(cluster_dict[key])
        else:
            preds.append([])

    return preds
```

```python
# map ids to x,y,z idea
# try to use and intermedate item to corraled the data together

# build models
sess = tf.InteractiveSession()
```

```python
class Model:
    def __init__(self,
                 sess,
                 insize,                 
                 stdev = 0.5):
        # inputs
        self.ident     = tf.placeholder(dtype=tf.float32, shape=[None,insize], name="x")
        self.contrast  = tf.placeholder(dtype=tf.float32, shape=[None,insize], name="c")
        self.other_xyz = tf.placeholder(dtype=tf.float32, shape=[None,3], name="other_xyz")
        self.exp_distance  = tf.placeholder(dtype=tf.float32, shape=[None,1], name="exp_distance")

        Wo = tf.Variable(tf.random_normal([insize,3], stddev=stdev, dtype=tf.float32))
            
        # note that the reality of this embedding is just a raw table lookup...
        self.xyz = tf.matmul(self.ident, Wo)

        xyz_mag       = tf.sqrt(tf.reduce_sum(tf.mul(self.xyz,       self.xyz),      1)) 
        other_xyz_mag = tf.sqrt(tf.reduce_sum(tf.mul(self.other_xyz, self.other_xyz),1)) 

        # magtuide err all vectors should be unit 1
        unitsize_delta = xyz_mag - 1
        unitsize_err = tf.mul(unitsize_delta, unitsize_delta)
        
        # angle error .. mapped point should be at correct angle from other ids xyz mapping
        cos_angle = tf.reduce_sum(tf.mul(self.xyz, self.other_xyz),1) / xyz_mag / other_xyz_mag       
        cos_ang_delta = self.exp_distance - cos_angle
        cos_ang_err = cos_ang_delta * cos_ang_delta
        
        # noise contrast error .. I should be as distance as possible from another random item in the space
        rand_xyz     = tf.matmul(self.contrast, Wo)
        rand_xyz_mag = tf.sqrt(tf.reduce_sum(tf.mul(rand_xyz, rand_xyz),1)) 
        rand_cos_ang = 1.0 + tf.reduce_sum(tf.mul(self.xyz, rand_xyz),1) / xyz_mag / rand_xyz_mag
        
        self.lossfunc = unitsize_err + cos_ang_err  + rand_cos_ang
                    
        self.train_op = tf.train.AdamOptimizer().minimize(self.lossfunc)

        sess.run(tf.initialize_all_variables())

    def run(self,
            sess,
            in_ident):
        return sess.run(self.xyz,feed_dict={self.ident: in_ident})
        
    def train(self,
              sess,
              in_ident,
              other_xyz,
              out_distance,
              NEPOCH = 10):                
        for i in range(NEPOCH):
            #in_contrast = np.random.rand(in_ident.shape[0],in_ident.shape[1])
            in_contrast = np.random.permutation(in_ident)
            fdict = {self.ident:        in_ident,
                     self.contrast:     in_contrast,
                     self.other_xyz:    other_xyz,  
                     self.exp_distance: out_distance}
            
            sess.run(self.train_op,feed_dict=fdict)
```

```python
def labels_classes(labels):
    # find the unique class labels
    return np.unique(labels)

def labels_to_one_hots(labels,classes):
    num_labels = labels.shape[0]
    num_classes = classes.shape[0]

    # create index of label into the hots..
    #   ie covert the label to a number indexing into classes
    label_index = np.zeros(labels.shape[0], dtype=int)
    idx = 0
    for c in classes:
        label_index[labels == c] = idx
        idx += 1

    # compute the offsets into the one_hots (if it was flat)
    hot_indexes = np.arange(num_labels) * num_classes + label_index

    # and then write each item that is a hot as a hot
    one_hots = np.zeros((num_labels, num_classes))
    one_hots.flat[hot_indexes] = 1

    return one_hots

def embed_setup(train_set):
     #extract data
    idx = ~ pd.isnull(train_set.orig_destination_distance)

    distance  = np.array(train_set.orig_destination_distance[idx])
    user_ids  = np.array(train_set.user_location_country[idx])
    hotel_ids = np.array(train_set.hotel_country[idx])
    
    user_classes  = labels_classes(np.sort(user_ids))
    hotel_classes = labels_classes(np.sort(hotel_ids))
    
    # assume distance was miles.. convert to cos(ang)
    distance =  np.reshape(distance, (distance.shape[0],1)) / 3959
    distance = np.cos(distance) 
    
    user_one_hots  = labels_to_one_hots(user_ids,user_classes)
    hotel_one_hots = labels_to_one_hots(hotel_ids,hotel_classes)

    user_model = Model(sess, user_classes.shape[0])
    hotel_model = Model(sess, hotel_classes.shape[0])

    return user_model, hotel_model, user_one_hots, hotel_one_hots, distance, user_classes,hotel_classes

def embed_train(sess, user_model, hotel_model, user_ids, hotel_ids, distance):    
    # start training cycle
    step = 2000
    rounds = user_ids.shape[0] / step + 1 # meh i know but its close enough..
    for c in range(100):
        for i in range(rounds):
            # first batch this up so my machine doesnt die..
            offset = i*step % user_ids.shape[0] 
            oend   = offset + step         
            if oend > user_ids.shape[0]:
                oend = user_ids.shape[0]
        
            batch_user     = user_ids[offset:oend]
            batch_hotel    = hotel_ids[offset:oend]
            batch_distance = distance[offset:oend]
        
            # predict xyz from one model 
            hotel_xyz = hotel_model.run(sess, batch_hotel)
        
            # then retrain other model againest xyz
            user_model.train(sess,
                             batch_user,
                             hotel_xyz,
                             batch_distance)
        
            # then swap over
            user_xyz = user_model.run(sess, batch_user)
        
            hotel_model.train(sess,
                              batch_hotel,
                              user_xyz,
                              batch_distance)

            # then check final error -- yes i know this is not the correct way to validate
            hotel_xyz = hotel_model.run(sess, batch_hotel)
    
            xyz_distance = user_xyz - hotel_xyz
            xyz_distance = np.sqrt(xyz_distance * xyz_distance)
        
            err = batch_distance - xyz_distance
            err = err * err
        
            print "err:", np.mean(err), " cycle:", c, " round:", i

def embed_map(sess, user_model, hotel_model, user_classes, hotel_classes): 
    user_xyz  = user_model.run(sess, np.eye(user_classes.shape[0]))
    hotel_xyz = hotel_model.run(sess, np.eye(hotel_classes.shape[0]))   
    
    figh = plt.figure(figsize=(15,15))
    ax3D = figh.add_subplot(111, projection='3d')
    
    #compute wireframe sphere
    u, v = np.mgrid[0:2*np.pi:20j, 0:np.pi:10j]
    x=np.cos(u)*np.sin(v)
    y=np.sin(u)*np.sin(v)
    z=np.cos(v)
    
    ax3D.plot_wireframe(x, y, z, color="r")
    ax3D.scatter(hotel_xyz[:,0], hotel_xyz[:,1], hotel_xyz[:,2], marker='o')  
    
    figu = plt.figure(figsize=(15,15))
    ax3D = figu.add_subplot(111, projection='3d')
    ax3D.plot_wireframe(x, y, z, color="r")
    ax3D.scatter(user_xyz[:,0], user_xyz[:,1], user_xyz[:,2], marker='o')  
    
    plt.show()
```
