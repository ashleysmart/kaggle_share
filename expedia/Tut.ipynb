```python
def segmentData(data):
    date_time = pd.to_datetime(data["date_time"])
    data["date_time"] = date_time
    data["year"]      = date_time.dt.year
    data["month"]     = date_time.dt.month
    
    train = data[((data.year == 2013) | ((data.year == 2014) & (data.month < 8)))]
    valid = data[((data.year == 2014) & (data.month >= 8))]
    valid = valid[valid.is_booking == True]
    
    return train, valid
```

```python
def make_key(items):
    return "_".join([str(i) for i in items])

def model1_train(match_cols, train_set):
    cluster_cols = match_cols + ['hotel_cluster']
    groups = train_set.groupby(cluster_cols)
    top_clusters = {}
    for name, group in groups:
        clicks = len(group.is_booking[group.is_booking == False])
        bookings = len(group.is_booking[group.is_booking == True])
    
        score = bookings + .15 * clicks
    
        clus_name = make_key(name[:len(match_cols)])
        if clus_name not in top_clusters:
            top_clusters[clus_name] = {}
        top_clusters[clus_name][name[-1]] = score

    cluster_dict = {}
    for n in top_clusters:
        tc = top_clusters[n]
        top = [l[0] for l in sorted(tc.items(), key=operator.itemgetter(1), reverse=True)[:5]]
        cluster_dict[n] = top

    return cluster_dict

def model1_run(match_cols, cluster_dict,valid_set):
    preds = []
    for index, row in valid_set.iterrows():
        key = make_key([row[m] for m in match_cols])
        if key in cluster_dict:
            preds.append(cluster_dict[key])
        else:
            preds.append([])

    return preds
```

```python
class Model:
    def __init__(self,
                 sess,
                 nhidden = 24,
                 stdev = 0.5):
        self.NHIDDEN = nhidden
        self.STDEV = stdev

        # inputs
        self.ident = tf.placeholder(dtype=tf.float32, shape=[None,1], name="x")
        self.other_xyz = tf.placeholder(dtype=tf.float32, shape=[None,3], name="other_xyz")
        self.exp_distance  = tf.placeholder(dtype=tf.float32, shape=[None,1], name="exp_distance")

        Wh = tf.Variable(tf.random_normal([1,self.NHIDDEN], stddev=self.STDEV, dtype=tf.float32))
        bh = tf.Variable(tf.random_normal([1,self.NHIDDEN], stddev=self.STDEV, dtype=tf.float32))
        Wo = tf.Variable(tf.random_normal([self.NHIDDEN,3], stddev=self.STDEV, dtype=tf.float32))
        bo = tf.Variable(tf.random_normal([1,3], stddev=self.STDEV, dtype=tf.float32))
            
        hidden_layer = tf.nn.relu(tf.matmul(self.ident, Wh) + bh)
        self.xyz = tf.matmul(hidden_layer,Wo) + bo
        self.distance = tf.abs(self.xyz - self.other_xyz)

        self.lossfunc = tf.square(self.distance - self.exp_distance)
            
        self.train_op = tf.train.AdamOptimizer().minimize(self.lossfunc)

        sess.run(tf.initialize_all_variables())

    def run(self,
            sess,
            in_ident):
        return sess.run(self.xyz,feed_dict={self.ident: in_ident})                                     
        
    def train(self,
              sess,
              in_ident,
              other_xyz,
              out_distance,
              NEPOCH = 1000):                
        loss = np.zeros(NEPOCH) # store the training progress here.
        for i in range(NEPOCH):
            fdict = {self.ident:        in_ident,
                     self.other_xyz:    other_xyz,  
                     self.exp_distance: out_distance}
            
            sess.run(self.train_op,feed_dict=fdict)
            
            #if i % 100:
            #    loss = sess.run(self.lossfunc, feed_dict=fdict)
            #    print np.sum(loss)
                
            # plot the error rates
            #plt.figure(figsize=(8, 8))
            #plt.plot(np.arange(100, NEPOCH,1), loss[100:], 'r-')
            # plt.show()
            
```

```python
# map ids to x,y,z idea
# try to use and intermedate item to corraled the data together

# build models
sess = tf.InteractiveSession()
```

```python
user_model = Model(sess)
hotel_model = Model(sess)
```

```python
def embedd_id_xyz(sess, user_model, hotel_model, train_set):    
     #extract data
    idx = ~ pd.isnull(train_set.orig_destination_distance)

    distance  = np.array(train_set.orig_destination_distance[idx])
    user_ids  = np.array(train_set.user_location_country[idx])
    hotel_ids = np.array(train_set.hotel_country[idx])
    
    user_ids  = np.reshape(user_ids,  (user_ids.shape[0],1))
    hotel_ids = np.reshape(hotel_ids, (hotel_ids.shape[0],1))
    distance =  np.reshape(distance, (distance.shape[0],1))

    # start training cycle
    for i in range(100):
        # first predict xyz from one model 
        step = 1000
        offset = i*step % user_ids.shape[0] 
        oend   = offset + step         
        if oend > user_ids.shape[0]:
            oend = user_ids.shape[0]
        
        batch_user     = user_ids[offset:oend]
        batch_hotel    = hotel_ids[offset:oend]
        batch_distance = distance[offset:oend]
        
        hotel_xyz = hotel_model.run(sess, batch_hotel)
        
        # then retrain other model againest xyz
        user_model.train(sess,
                         batch_user,
                         hotel_xyz,
                         batch_distance)
        
        # then swap over
        user_xyz = user_model.run(sess, batch_user)
        
        hotel_model.train(sess,
                          batch_hotel,
                          user_xyz,
                          batch_distance)

        # then check final error
        hotel_xyz = hotel_model.run(sess, batch_hotel)
    
        xyz_distance = user_xyz - hotel_xyz
        xyz_distance = np.sqrt(xyz_distance * xyz_distance)
        
        err = batch_distance - xyz_distance
        err = err * err
        
        print "err:", np.mean(err)
        
```
